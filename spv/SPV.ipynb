# Sentinel Parcel Viewer

### Download dependencies and set up the working environment ###

```python
import urllib.request, runpy

url = "https://raw.githubusercontent.com/ec-jrc/cbm/refs/heads/main/spv/bootstrap.py"
urllib.request.urlretrieve(url, "bootstrap.py")

_ = runpy.run_path("bootstrap.py")
```

### Authenticate ###

```python
#connection to openEO backend
import openeo
connection = openeo.connect(url="openeofed.dataspace.copernicus.eu/openeo/")
connection.authenticate_oidc(store_refresh_token=True)
```

### Loading interface ###

```python
import openeo_utils as ou
ui = ou.getInterface(connection)
display(ui)
```
